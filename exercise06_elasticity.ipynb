```python
# @title 'PLAY' icon here below
def get_valid_input():
    while True:
        ID = input("Enter a number (last 3 digits of your student ID): ")

        # Check if the entered number has exactly three digits
        if len(ID) != 3 or not ID.isdigit():
            print("Invalid input! Please enter exactly three digits.")
        else:
            return int(ID)

def main():
    while True:
        # Get valid input from the user
        ID = get_valid_input()

        # Convert the number to a string, reverse it, and convert it back to an integer
        padded_ID = str(ID).zfill(3)
        RDI = int(padded_ID[::-1])

        # Print the original and reversed numbers
        print("Last three digits of Student ID:", ID)
        print("Reverse digits of Student ID:", RDI)

        # Compare ID and RDI to find the greater and smaller values
        Greater = max(ID, RDI)
        Smaller = min(ID, RDI)

        """print("\nPART 1")
        print("     Task 1: calculating opportunity cost of working in the high-paid jobs")
        ValueA = ((Smaller / 9) * 1.5)
        # Format the output to show two decimal places after calculation
        ValueA_formatted = f"{ValueA:.2f}"
        print("         ValueA:", ValueA_formatted)

        print("     Task 2: calculating opportunity cost of working in the low-paid jobs")
        ValueB = (Greater / 9)
        # Format the output to show two decimal places after calculation
        ValueB_formatted = f"{ValueB:.2f}"
        print("         ValueB:", ValueB_formatted)

        # Convert formatted strings back to float for calculation
        ValueA = float(ValueA_formatted)
        ValueB = float(ValueB_formatted)"""

        print("\nPART 1")
        print("     Task 1: calculating opportunity cost of working in the high-paid jobs")
        ValueA = round((Smaller / 9) * 1.5)
        ValueA = ValueA if ValueA != 0 else 1  # Ensure it is not zero
        print("         ValueA:", ValueA)

        print("     Task 2: calculating opportunity cost of working in the low-paid jobs")
        ValueB = round(Greater / 9)
        ValueB = ValueB if ValueB != 0 else 1  # Ensure it is not zero
        print("         ValueB:", ValueB)


        print("\nPART 2")
        print("         what is the change in the income (ΔI) from switching the job ?")
        print("         ΔI = ((HighSalary - LowSalary)/LowSalary)*100%)")

        # Compare ValueA and ValueB to find Last and Prev
        Last = max(ValueA, ValueB)
        Prev = min(ValueA, ValueB)



        # Calculate I
        I = round(((Last - Prev) / Prev) * 100)

        # Print the values
        print("         ΔI:", I, "%")
        print("     Task 1: what will be the % change in the number of smartphones demanded?")
        print("         Einc = ΔQsmart / ΔI; then ΔQsmart = Einc * ΔI")
        print("         Einc = 1 + 0.XX, where XX last 2 digits of your ID")

        # Extract the last two digits
        last_two_digits = ID % 100

        # Convert the last two digits to a string
        last_two_digits_str = str(last_two_digits).zfill(2)

        # Construct the result
        Einc = "1." + last_two_digits_str

        print("         Einc = ", Einc)
        # Calculate Qsmart
        Qsmart = float(Einc) * I

        # Round the result to the first decimal place
        Qsmart_rounded = round(Qsmart, 1)

        # Print the rounded result
        print("         ΔQsmart = ", Qsmart_rounded, "%")
        print("     Task 2: what is the new number of smartphones demanded?")
        print("         NewQsmart = OldQsmart (defined by your ID) + change")
        print("         change = OldQsmart*(ΔQsmart/100)")
        OldQsmart = ID
        NewQsmart = round(OldQsmart + OldQsmart * (Qsmart / 100))

        # Format New quantity of smartphones with commas
        formatted_NewQsmart = "{:,}".format(NewQsmart)
        print("         NewQsmart = ", formatted_NewQsmart)

        print("\nPART 3")
        print("         what is the cross-price elasticity of demand for smartphones?")
        print("         Ecross = 3 - Einc, where Einc income elasticity of demand for smartphones")
        Ecross = 3 - float(Einc)
        print("         Ecross = ", Ecross)

        print("     Task 1: calculating % change in the price of laptops")
        print("         ΔPlaptops: Ecross = ΔQsmart /ΔPlaptops; ΔPlaptops = ΔQsmart / Ecross")

        ΔPlaptops = Qsmart / Ecross
        # Round the result to the first decimal place
        ΔPlaptops_rounded = round(ΔPlaptops, 1)
        print("         ΔPlaptops = ", ΔPlaptops_rounded, "%")

        print("     Task 2: What will be the new price for laptops?")
        print("         NewPlaptop = OldPlaptop ($450) + change (ΔPlaptops)")
        print("         change = OldPlaptop*(ΔPlaptops/100)")
        OldPlaptop = 450
        NewPlaptop = round(OldPlaptop + OldPlaptop * (ΔPlaptops / 100))

        # Format Price of laptops with commas
        formatted_NewPlaptop = "{:,}".format(NewPlaptop)
        print("         NewPlaptop = ", formatted_NewPlaptop)

        print("\nPART 4")
        print("         what is the elasticity of demand for laptops?")
        print("         Edemlap = 0.5 + Ecross, where Ecross crossprice elasticity of demand for smartphone")
        Edemlap = 0.5 + Ecross
        print("         Edemlap = {:.2f}".format(Edemlap))

        print("     Task 1: find out the % change in the number of laptops demanded")
        print("         ΔQlaptops: Edemlap = ΔQlaptops /ΔPlaptops; ΔQlaptops = ΔPlaptops * Edemlap")

        ΔQlaptops = ΔPlaptops * Edemlap

        # Round the result to the first decimal place
        ΔQlaptops_rounded = round(ΔQlaptops, 1)
        print("         ΔQlaptops = ", ΔQlaptops_rounded, "%")

        print("     Task 2: What will be the new number of laptops demanded?")
        print("         NewQlaptop = OldQlaptop (the last 3 digits of your ID)", RDI, " + change (ΔQlaptops)")
        print("             change = OldQlaptop*(ΔQlaptops/100)")

        OldQlaptop = RDI
        NewQlaptop = round(OldQlaptop + OldQlaptop * (ΔQlaptops / 100))
        # Format Quantity of laptops with commas
        formatted_NewQlaptop = "{:,}".format(NewQlaptop)
        print("         NewQlaptop = ", formatted_NewQlaptop)

        print("\nPART 5")

        print("     Task 1: calculate the change in the total revenue for laptops")

        print("         OldRevenue = OldPlaptop * OldQlaptop")

        OldRevenue = OldPlaptop * OldQlaptop
        # Format OldRevenue with commas
        formatted_old_revenue = "{:,}".format(OldRevenue)
        # Print the formatted OldRevenue
        print("             OldRevenue =", formatted_old_revenue)

        print("         NewRevenue = NewPlaptop * NewQlaptop")
        NewRevenue = NewPlaptop * NewQlaptop
        # Format NewRevenue with commas
        formatted_new_revenue = "{:,}".format(NewRevenue)
        # Print the formatted NewRevenue
        print("             NewRevenue =", formatted_new_revenue)

        print("         ΔRevenue = NewRevenue - OldRevenue")
        Revenue = NewRevenue - OldRevenue
        # Format Revenue with commas
        formatted_Revenue = "{:,}".format(Revenue)
        print("             ΔRevenue = ", formatted_Revenue)

        print("     Task 2: determine whether there was a fall or growth in the total revenue for laptops?")
        if NewRevenue > OldRevenue:
            print("         There was GROWTH in the total revenue")
        elif NewRevenue == OldRevenue:
            print("         The TOTAL revenue has NOT changed")
        else:
            print("         There was FALL in the total revenue")

        # Ask if the user wants to enter another number
        again = input("\nWould you like to enter another number? (y/n): ").lower()
        if again != 'y':
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```
